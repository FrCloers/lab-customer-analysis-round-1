Read the three files into python as dataframes

```python
# Reading data
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')  # Show them to read and concat 2 files , other two will be done by students
```

Rearrange the columns in the dataframe as needed

```python
cols1=[]
for c in file1.columns:
    cols1.append(c.lower())

file1.columns=cols1

# Similarly for the other dataframe:
cols2=[]
for c in file2.columns:
    cols2.append(c.lower())

file2.columns=cols2

cols3=[]
for c in file3.columns:
    cols3.append(c.lower())

file3.columns=cols3
```

```python
data1 = file1[['customer', 'st', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
data1.head()
```

```python
data2=file2[['customer', 'st', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
data2.head()
```

```python
data3 = file3.rename(columns={'state':'st'})
data3.head()
```

```python
data3=data3[['customer', 'st', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
data3.head()
```

Concatenate the three dataframes

```python
column_names = file1.columns # We store the column names of the first file into a variable
data = pd.DataFrame(columns=column_names) # We create a new dataframe variable with the same columnames as the first file
data = pd.concat([data,data1, data2,data3], axis=0) # Axis 0 means contatenation by row.
data.shape # This is to know the (number_of_rows, number_of_columns) of the new dataframe "data"
```
